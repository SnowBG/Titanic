```python
import pandas as pd
```

```python
df = pd.read_csv('c:/p/Titanic/train.csv', header=0)
```

```python
print(df.info())
```

```python
df.head()
```

```python
#remove not helpful informations:
cols = ['Name','Ticket','Cabin']
df = df.drop(cols, axis=1)
df.head(10)
```

```python
#remove all rows that has missing values (NaN):
'''Not good. We lose some values.'''
'''df = df.dropna()
df.head(10)'''
```

```python
df.shape
```

```python
#Now we convert the Pclass, Sex, Embarked to columns in pandas and drop them after conversion.

dummies = []
cols = ['Pclass','Sex','Embarked']
[dummies.append(pd.get_dummies(df[col])) for col in cols]
```

```python
titanic_dummies = pd.concat(dummies, axis=1)
```

```python
titanic_dummies.head(10)
```

```python
#finally we concatenate to the original dataframe columnwise

df = pd.concat((df,titanic_dummies),axis=1)
```

```python
#Now that we converted Pclass, Sex, Embarked values into columns, we drop the redundant same columns from the dataframe
df = df.drop(['Pclass','Sex','Embarked'], axis =1)
df.head(20)
```

```python
df.shape
```

```python
df.info()
```

```python
#Here we interpolate the missing values from the AGE column:
df['Age'] = df['Age'].interpolate()
df.info()
```
